```python
import warnings
from rply import ParserGenerator, LexerGenerator, ParsingError
from rply.lexer import LexingError
from rply.token import BaseBox

# cria um objeto LexerGenerator
lg = LexerGenerator()

# insere as regras de formação (expressões regulares) dos tokens da linguagem

#PREDEFINED IDENTIFIERS
lg.add('VOID', r'void')


lg.add('BOOLEAN', r'boolean')
lg.add('TRUE', r'true')
lg.add('FALSE', r'false')
lg.add('READ', r'read')
lg.add('INTEGER', r'integer')
lg.add('WRITE', r'write')
lg.add('LABELS', r'labels')
lg.add('TYPES', r'types')
lg.add('VARS', r'vars')
lg.add('VAR', r'var')
lg.add('GOTO', r'goto')
lg.add('RETURN', r'return')
lg.add('IF', r'if')
lg.add('ELSE', r'else')
lg.add('WHILE', r'while')


lg.add('FUNCTIONS', r'functions')

lg.add('EQUAL', r'\==')
lg.add('ATTR', r'\=')

lg.add('NOTEQUAL', r'!=')
lg.add('NOT', r'\!')

lg.add('GREATEREQUALTHAN', r'>=')
lg.add('GREATERTHAN', r'>')

lg.add('LESSEREQUALTHAN', r'<=')
lg.add('LESSERTHAN', r'<')

lg.add('OR', r'\|\|')
lg.add('AND', r'\&\&')


lg.add('NUMBER', r'\d+')
lg.add('PLUS', r'\+')
lg.add('MINUS', r'-')
lg.add('MULT', r'\*')
lg.add('DIV', r'/')
lg.add('LPAREN', r'\(')
lg.add('RPAREN', r'\)')
lg.add('LCHAVE', r'\{')
lg.add('RCHAVE', r'\}')
lg.add('COMMA', r'\,')
lg.add('SEMICOLON', r'\;')
lg.add('COMMA', r'\,')
lg.add('COLON', r'\:')
lg.add('LBRACKET', r'\[')
lg.add('RBRACKET', r'\]')
lg.add('COLON', r'\:')



lg.add('ID', r'[a-z][a-z0-9]*')




lg.add('MOD', r'%')

# cria uma regra para ignorar caracteres de espaços
lg.ignore('\s+')
lg.ignore('\#.+')

#####################################################
pg = ParserGenerator(
        ['VOID', 'ID', 'LPAREN', 'RPAREN', 'LCHAVE', 'RCHAVE', 'VAR','COLON', 'COMMA', 'ATTR',
         'MULT', 'DIV', 'AND', 'PLUS', 'MINUS', 'OR','INTEGER','NUMBER','NOT','COLON','VAR',
         'COMMA','SEMICOLON', 'GREATEREQUALTHAN', 'LESSEREQUALTHAN','LESSERTHAN','GREATERTHAN','EQUAL','NOTEQUAL',
         'FUNCTIONS','LABELS','VARS','RETURN','IF','ELSE','WHILE', 'GOTO','WRITE', 'READ'],
     )

@pg.production('program : function')
def program(p):
    return p[0].getstr()

@pg.production('function : VOID ID formal_parameters block')
@pg.production('function : ID ID formal_parameters block')
def function(p):

    return p[0]

@pg.production('formal_parameters : LPAREN formal_parameter RPAREN')
@pg.production('formal_parameters : LPAREN  RPAREN')
def formal_parameters(p):
    return p[0]

@pg.production('formal_parameter :  expression_parameter formal_parameter1')
@pg.production('formal_parameter :  expression_parameter')
def formal_parameter(p):
    return p[0]

@pg.production('formal_parameter1 :  SEMICOLON formal_parameter')
def formal_parameter1(p):
    return p[0]

@pg.production('expression_parameter :     identifier_list COLON ID ')
@pg.production('expression_parameter : VAR identifier_list COLON ID ')
def expression_parameter(p):
    return p[0]

@pg.production('identifier_list : ID ')
@pg.production('identifier_list : ID identifier_list1')
def identifier_list(p):
    return p[0]

@pg.production('identifier_list1 : COMMA identifier_list')
def identifier_list1(p):
    return p[0]

@pg.production('block : labels variables funcao body')
@pg.production('block : labels variables  body')
@pg.production('block : labels  funcao body')


@pg.production('block : variables funcao body')
@pg.production('block : variables body')
@pg.production('block : funcao body')

@pg.production('block : body')
def block(p):
    return p[0]

@pg.production('funcao : FUNCTIONS function')
@pg.production('funcao : FUNCTIONS function funcao1')
def funcao(p):
    return p[0]

@pg.production('funcao1 :  function ')
@pg.production('funcao1 :  function funcao1')
def funcao1(p):
    return p[0]



@pg.production('labels : LABELS identifier_list SEMICOLON')
def labels(p):
    return p[0]

@pg.production('variables : VARS identifier_list COLON type SEMICOLON')
@pg.production('variables : VARS identifier_list COLON type SEMICOLON variables1')
def variables(p):
    return p[0]
@pg.production('variables1 :  identifier_list COLON type SEMICOLON ')
@pg.production('variables1 :  identifier_list COLON type SEMICOLON variables1  ')
def variables1(p):
    return p[0]

@pg.production('type : ID')
def type(p):
    return p[0]

@pg.production('body : LCHAVE body1  RCHAVE')
@pg.production('body : LCHAVE  RCHAVE')
def body(p):
    return p[0]

@pg.production('body1 : statement body1 ')
@pg.production('body1 : statement ')
def body1(p):
    return p[0] 

   

@pg.production('statement : ID COLON unlabeled_statement')
@pg.production('statement : unlabeled_statement')
@pg.production('statement : compound')
def statement(p):
    return p[0]

@pg.production('compound : LCHAVE  unlabeled_statement RCHAVE')
@pg.production('compound : LCHAVE  unlabeled_statement compound1 RCHAVE')
def compound(p):
    return p[0]
@pg.production('compound1 : unlabeled_statement compound1  ')
@pg.production('compound1 : unlabeled_statement ')
def compound1(p):
    return p[0]

@pg.production('unlabeled_statement : assignment ')
@pg.production('unlabeled_statement : function_call_statement')
@pg.production('unlabeled_statement : vaipara')
@pg.production('unlabeled_statement : retorno')
@pg.production('unlabeled_statement : conditional')
@pg.production('unlabeled_statement : repetitive')
@pg.production('unlabeled_statement : empty_statement')
def unlabeled_statement(p):
    return p[0]


@pg.production('empty_statement :  SEMICOLON')
def empty_statement(p):
    return p[0]

@pg.production('vaipara : GOTO ID SEMICOLON')
def vaipara(p):
    return p[0]

@pg.production('retorno : RETURN SEMICOLON')
@pg.production('retorno : RETURN expression SEMICOLON')
def retorno(p):
    return p[0]

    
@pg.production('repetitive : WHILE LPAREN expression RPAREN compound')
def repetitive(p):
    return p[0]

@pg.production('conditional : IF LPAREN expression RPAREN compound')
@pg.production('conditional : IF LPAREN expression RPAREN compound ELSE compound')
def conditional(p):
    return p[0]

@pg.production('function_call_statement : function_call SEMICOLON')
@pg.production('function_call_statement : WRITE LPAREN  RPAREN SEMICOLON')
@pg.production('function_call_statement : WRITE LPAREN expression_list RPAREN SEMICOLON')
@pg.production('function_call_statement : READ LPAREN  RPAREN SEMICOLON')
@pg.production('function_call_statement : READ LPAREN expression_list RPAREN SEMICOLON')
def function_call_statement(p):
    return p[0]


    
@pg.production('assignment : variable ATTR expression  SEMICOLON')
def assignment(p):
    return p[0]

    
@pg.production('variable : ID ')
def variable(p):
    return p[0]

    
@pg.production('expression : simple_expression ')
@pg.production('expression : simple_expression relational_operator simple_expression ')
def expression(p):
    return p[0]

@pg.production('relational_operator :   EQUAL')
@pg.production('relational_operator :   NOTEQUAL')
@pg.production('relational_operator :   LESSERTHAN')
@pg.production('relational_operator :   LESSEREQUALTHAN')
@pg.production('relational_operator :   GREATEREQUALTHAN')
@pg.production('relational_operator :   GREATERTHAN')
def relational_operator(p):
    return p[0]

@pg.production('simple_expression :   PLUS term ')
@pg.production('simple_expression :   PLUS term additive_operator term ')
@pg.production('simple_expression :   MINUS term additive_operator term')
@pg.production('simple_expression :   term additive_operator term')
@pg.production('simple_expression :   term additive_operator term')
@pg.production('simple_expression :   term ')
@pg.production('simple_expression :   MINUS term ')
def simple_expression(p):
    return p[0]

@pg.production('additive_operator :   PLUS')
@pg.production('additive_operator :   MINUS')
@pg.production('additive_operator :   OR')
def additive_operator(p):
    return p[0]

@pg.production('term :  factor ')
@pg.production('term :  factor multiplicative_operator factor')
def term(p):
    return p[0]

@pg.production('multiplicative_operator :  MULT')
@pg.production('multiplicative_operator :  DIV')
@pg.production('multiplicative_operator :  AND')
def multiplicative_operator(p):
    return p[0]

@pg.production('factor :  variable')
@pg.production('factor :  NUMBER')
@pg.production('factor :  function_call')
@pg.production('factor :  LPAREN expression RPAREN')
@pg.production('factor :  NOT factor')
def factor(p):
    return p[0]

@pg.production('function_call : ID LPAREN expression_list RPAREN ')
@pg.production('function_call : ID LPAREN  RPAREN ')
def function_call(p):
    return p[0]

@pg.production('expression_list : expression ')
@pg.production('expression_list : expression expression_list1')
def expression_list(p):
    return p[0]

@pg.production('expression_list1 :  COMMA expression_list')
def expression_list1(p):
    return p[0]




```
